```python
import numpy as np
import pandas as pd
df=pd.DataFrame()
```

```python
data=[['seasonal flu',128583],['maternal deaths',81734],['HIV/AIDS',44533],['cancer',2171792],['malaria',259381],['smoking',1321972],['alcohol',661381],['suicide',283569],['road accidents',356963]]
```

```python
df = pd.DataFrame(data, columns = ['Death', 'Numbers']) 
```

```python
df['NumbersYesterday']=df['Numbers']/96.5
```

```python
df
```

```python
dfyday=df.sort_values(by=['Numbers'],ascending=False)[['Death','NumbersYesterday']]
```
