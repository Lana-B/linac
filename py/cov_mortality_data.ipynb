```python
data=[['seasonal flu',425894],['maternal deaths',268990],['HIV/AIDS',1462876],['cancer',7147458],['malaria',853633],['smoking',4350511],['alcohol',2176629],['suicide',933234],['road accidents',1174777]]
```
